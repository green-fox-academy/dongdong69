# 锐化

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
import numpy as np
 
def custom_blur_demo(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]], np.float32) #锐化
    dst = cv.filter2D(image, -1, kernel=kernel)
   # cv.imshow("custom_blur_demo", dst)
    return dst

plt.figure(num='astronaut',figsize=(15,15))  #创建一个名为astronaut的窗口,并设置大小 

src = cv.imread("Aragaki_Yui.jpg")

dst = custom_blur_demo(src)

plt.subplot(1,2,1)
plt.imshow(src)
plt.title("input")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(dst)
plt.title('sharp')
plt.axis('off')
plt.show()
```
