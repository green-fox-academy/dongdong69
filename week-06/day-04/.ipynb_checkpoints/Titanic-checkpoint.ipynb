```python
import pandas as pd
import numpy as np
import math
```

```python
train = pd.read_csv('train.csv')
```

```python
def haveCabin(value):
    try:
        if math.isnan(value):
            return 0
    except:
        return 1

train['Cabin_have'] = train['Cabin'].apply(haveCabin)
```

```python
train.Age=train.Age.fillna(train.Age.mean())

def age_range(age):
    if math.isnan(age):
        return 0
    elif age <= 5:
        return 1
    elif age <= 15:
        return 2
    elif age <= 30:
        return 3
    elif age <= 45:
        return 4
    elif age <= 60:
        return 5
    else:
        return 6
        
#train['Age_range'] = train['Age'].apply(age_range)
```

```python
def sex_label(sex):
    if sex == 'male':
        return 1
    else:
        return 0
    
train['Sex_label'] = train['Sex'].apply(sex_label)
```

```python
train.info()
```

```python
train.sample(10)
```

```python
def embarked_label(embark):
    if embark == 'C':
        return 0
    elif embark == 'Q':
        return 1
    elif 'S':
        return 2
    
train['Embarked_label'] = train['Embarked'].apply(embarked_label)
```

```python
train.sample(10)
```

```python
def SibSp_label(sibsp):
    if sibsp == 0:
        return 0
    elif sibsp <= 2:
        return 1
    else:
        return 2
    
train['SibSp_label'] = train['SibSp'].apply(SibSp_label)
```

```python
train.sample(10)
```

```python
def parch_label(parch):
    if parch == 0:
        return 0
    else:
        return 1
    
train['Parch_label'] = train['Parch'].apply(parch_label)
```

```python
train.sample(10)
```

```python
#dataframe for training
p_train = train[['Survived', 'Pclass', 'Cabin_have', 'Age', 'Sex_label', 'Embarked_label', 'SibSp_label', 'Parch_label']]
```

```python
train_ = np.array(p_train)
```

```python
train_Y = train_[:,0]
train_X = train_[:,1:]
```

```python
X_normed = train_X / train_X.max(axis=0)
X_normed
```

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(n_iter=40, eta0=0.1, random_state=0)
ppn.fit(X_normed, train_Y)
```

```python
test = pd.read_csv('test.csv')

test['Cabin_have'] = test['Cabin'].apply(haveCabin)

test.Age=test.Age.fillna(train.Age.mean())
#test['Age_range'] = test['Age'].apply(age_range)

test['Sex_label'] = test['Sex'].apply(sex_label)
test['Embarked_label'] = test['Embarked'].apply(embarked_label)
test['SibSp_label'] = test['SibSp'].apply(SibSp_label)
test['Parch_label'] = test['Parch'].apply(parch_label)

test.sample(10)
```

```python
p_test = test[['PassengerId','Pclass','Cabin_have', 'Age', 'Sex_label', 'Embarked_label', 'SibSp_label', 'Parch_label']]
test_ = np.array(p_test)
test_id = test_[:,0]
test_X = test_[:,1:]
max_m = test_X.max(axis=0)
X_normed_test = test_X / max_m
```

```python
test[test['Fare'].isnull()]
```

```python
y_hat = ppn.predict(X_normed_test)
#accuracy_score(y_test, y_pred)
```

```python
result = np.ones((len(test_id), 2))
result[:,0] = test_id
result[:,1] = y_hat
```

```python
def save_result(file_name, result):
    data = pd.DataFrame(result, columns=['PassengerId', 'Survived'])
    data['PassengerId'] = data['PassengerId'].astype('int64')
    data['Survived'] = data['Survived'].astype('int64')
    data.to_csv(file_name, index = False)
```

```python
save_result('logistic_regression3.csv', result)

##0.68 version 1
##0.64114 version 2
```

# Vote by logR, 3NN and DT

```python
knn_3 = pd.read_csv('knn_3.csv')
logR = pd.read_csv('logistic_regression3.csv')
DT = pd.read_csv('Decision_Tree.csv')
```

```python
vote = np.zeros((418, 3))
vote[:,0] = knn_3.Survived
vote[:,1] = logR.Survived
vote[:,2] = DT.Survived
```

```python
votes = vote.sum(axis=1)
for index, n in enumerate(votes):
    if n <= 1:
        votes[index] = 0
    else:
        votes[index] = 1

votes
```

```python
result = np.ones((len(test_id), 2))
result[:,0] = test_id
result[:,1] = votes
save_result('vote.csv', result)
## 0.78947
```
