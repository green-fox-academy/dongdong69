````python
#What is the average rating of apps in the store?
google_data.Rating.mean()
````

````python
#What is the ratio of free vs paid?
google_data[google_data.Type == 'Paid'].Type.count()/google_data[google_data.Type == 'Free'].Type.count()
````

````python
#What is the highest amount of Reviews an app received? And the lowest?
int_reviews = []
for review in google_data.Reviews:
    if 'M' in review:
        int_review = float(review[:review.find('M')]) * 1000000
    else:
        int_review = int(review)
    int_reviews.append(int_review)
print(max(int_reviews))
print(min(int_reviews))
````

````python
import time
timestring = google_data.sample()['Last Updated'][0]
time.mktime(time.strptime(timestring, '%B %d, %Y'))
````

````python
#What are the ten most recently updated apps?
time_list = []

for update_time in google_data['Last Updated']:
    try:
        int_time = time.mktime(time.strptime(update_time, '%B %d, %Y'))
        time_list.append(int_time)
    except:
        continue
        
google_data[[time.mktime(time.strptime(update_time, '%B %d, %Y')) == max(time_list)\
            for update_time in google_data['Last Updated']]]
````

````python
#Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.
rating_25 = google_data.sort_values(by=['Rating'], ascending=False).reset_index()[:int(0.25 * google_data.count()[0])]

end = time.mktime(time.strptime('August 8, 2018', '%B %d, %Y'))
start = time.mktime(time.strptime('June 8, 2018', '%B %d, %Y'))

rating_25[[time.mktime(time.strptime(update_time, '%B %d, %Y')) > start\
           for update_time in rating_25['Last Updated']]]
````

````python
#Which app not to download? - 
#Find an app which takes the most space on your phone, but is rated poorly (define poorly)?

# rating < 3.0 is a poor apps
poor_apps = google_data[google_data['Rating'] < 3.0]

max_size_byte = 0
for size in poor_apps['Size']:
    if 'M' in size:
        size_in_byte = float(size[:size.find('M')]) * 1000000
    if 'k' in size:
        size_in_byte = float(size[:size.find('k')]) * 1000
    if max_size_byte > size_in_byte:
        continue
    else:
        max_size_byte = size_in_byte
        max_size = size

not_to_download = poor_apps[poor_apps['Size'] == max_size]
not_to_download
````

````python
#drop the noise
google_data = google_data.drop(["Life Made WI-Fi Touchscreen Photo Frame"])
````

````python
#Create a reviews reliability of reviews index, which takes into account, 
#how many reviews does the app have and how high the rating is. What metric do you take? 
#Create as a new field.
google_data['Rating'].mean()
google_data[google_data['Rating'] < 2]

def reliability_reviews(value):
    record = np.where(int(value) > 50, True, False)
    return record

def reliability_rating(value):
    record = np.where(value > 1.0 and value != 5.0, True, False)
    return record

google_data["reliability"] = google_data["Rating"].apply(reliability_rating) & google_data['Reviews'].apply(reliability_reviews)
````
