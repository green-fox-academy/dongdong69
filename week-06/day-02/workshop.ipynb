# Request
---
The Requests library provides a simple tool to execute API HTTP methods without custom coding.

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize

BaseUrl = 'https://api.skypicker.com/flights?flyFrom=BRS&to=PRG&dateFrom=19/12/2019&dateTo=24/12/2019&partner=picky'
GettingData = requests.get(BaseUrl)

json_data = json.loads(GettingData.text)
```

```python
json_data['data']
df = json_normalize(json_data['data'])
```

```python
df
```

```python
df.sort_values(by=['price'], ascending=False).reset_index().loc[1]['route']
```

```python
df.info()
```

```python
def change(value):
    if len(value) == 1:
        return False
    else:
        return True
    
df['changes'] = df['route'].apply(change)
```

```python
flight = df[df['changes'] == False].sort_values(by = ['price']).reset_index().loc[1]
```
