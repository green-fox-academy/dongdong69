## Exam performance

```python
tsvr = pd.read_csv("exams.tsv", sep='\t', index_col='user')
tsv_array = tsvr.values
time_array = tsv_array[:,1]

tsv_array[:, 1] = normal_time_in_min(time_array)

highest = 0
highest_point = 0
for i in range(10):
    point, time = float(tsv_array[i, 0]), tsv_array[i, 1]
    if time:
        point_per_min = point/time
    else:
        point_per_min = 0
    
    print(f"user{i + 1}, his score is {point_per_min}")
    
    highest = [highest, i][point_per_min > highest_point]
    highest_point = [highest_point, point_per_min][point_per_min > highest_point]
    
print(f"the highest user is {highest + 1} and his score is {highest_point}")

def normal_time_in_min(time_array):
    index = 0
    
    for time in time_array:
        if time != time:
            total_mins = 0
            time_array[index] = total_mins
            index += 1
            continue
            
        index_h = time.find('h')
        index_m = time.find('m')
        index_s = time.find('s')
        
        #h to m
        if index_h is -1:
            hours = 0
        else:
            hours = float(time[:index_h])
        
        if index_m is -1:
            mins = 0
        else:
            mins = float(time[index_h + 1:index_m])
            
        if index_s is -1:
            secs = 0
        else:
            secs = float(time[index_m + 1:index_s])
            
        if index_h is -1 and index_m is -1 and index_s is -1:
            time_split = time.split(':')
            hours, mins, secs = float(time_split[0]), float(time_split[1]), float(time_split[2])
                          
            
        total_mins = hours*60 + mins + secs/60
        time_array[index] = total_mins
        index += 1
        
    return time_array
```

```python
csvr = open("users2.csv", "r")

Dict_lines = csv.DictReader(csvr)
users_list = []

for line in Dict_lines:
    dic_user = {}
    for key, value in line.items():
        dic_user[key] = value
    users_list.append(dic_user)

csvr.close()

j = json.dumps(users_list)

jsonw = open("users.json", encoding="utf-8", mode="w")
jsonw.write(j)
jsonw.close()
```

```python
from xml.dom.minidom import Document

with open("flowers.json", "r") as jsonr:
    json_dic = json.loads(jsonr.read())
    print(json_dic) 

doc = Document()
root = doc.createElement("flowers")
doc.appendChild(root)


#tree = ElementTree(root)

for flower in json_dic:
    elem_flower = doc.createElement('flower')

    root.appendChild(elem_flower)

    for key, val in flower.items():
        
        child = doc.createElement(key)
        child_text = doc.createTextNode(str(val))
        
        child.appendChild(child_text)
        elem_flower.appendChild(child)


with open("flowers.xml", 'w') as f:
    f.write(doc.toprettyxml())
```
