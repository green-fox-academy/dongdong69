```python
import csv
import matplotlib.pyplot as plt
```

```python
def reader():
    csvr = open('logs.csv', encoding="utf-8", mode="r")

    records_dic_list = []
    lines = csv.reader(csvr)
        
    for line in lines:
        record_dic = {}
        #the number of the line which can be useful when loading information
        record_dic["number"] = line[0]
        #the date
        record_dic["date"] = line[1][:10]
        #the specific time door opens
        record_dic["time"] = time_to_sec(line[1][12:])
        #the name of the door
        record_dic["door"] = line[5]
        #the name of stuff
        record_dic["name"] = line[7]
        #the id of stuff
        record_dic["id"] = line[12]
        #add into the recordsList
        records_dic_list.append(record_dic)

    return(records_dic_list)

def time_to_sec(time):
    ##time to sec : 09:00:00 to 32400
    time_list = time.split(':')
    h, m, s = time_list[0], time_list[1], time_list[2]

    return float(h)*3600 + float(m)*60 + float(s)
```

```python
def first_arrive(records):
    #recording the first arrival time for different in different day
    first_arrive_dic = {}
    for record in records:
        if record["name"] not in first_arrive_dic:
            first_arrive_dic[record["name"]] = {record["date"] : [record["time"], record["door"][:8]]}
        elif record["date"] not in first_arrive_dic[record["name"]]:
            first_arrive_dic[record["name"]][record["date"]] = [record["time"], record["door"][:8]]

    return first_arrive_dic
```

```python
#the records list which contain the information about
#
#number : string
#date : string
#specific time : float
#door : string
#name : string
#id : string
records_dic_list = reader()

# the set of doors
doors = set()
dates = set()
for record in records_dic_list:
    dates.add(record["date"])
    doors.add(record["door"])

#the id of different doors
door_dic = {}
for door in doors:
    door_dic[door[:8]] = door[8:]

print(door_dic)

#the record of the different time of different people's first arrive
first_arrive_record = first_arrive(records_dic_list)
print(first_arrive_record['Czender András'])

#there are 193 different people in Green Fox
#print(len(first_arrive_record))
```

```python
print(len(dates))
print(len(first_arrive_record['Czender András']))
```

```python
date_number_people = {}
for people in first_arrive_record:
    for date in first_arrive_record[people]:
        if date not in date_number_people:
            date_number_people[date] = 1
        else:
            date_number_people[date] += 1
```

```python
date_people_should_come = []
for date in date_number_people:
    if date_number_people[date] > 30:
        date_people_should_come.append(date)

print(date_number_people)
print(date_people_should_come)
```

```python
## how many skip days
people_coming_days = {}
for people in first_arrive_record:
    people_coming_days[people] = 0
    for date in first_arrive_record[people]:
        if date in date_people_should_come:
            people_coming_days[people] += 1
            
print(people_coming_days)
```

```python
first_arrive_record_should_come = {}
for people in first_arrive_record:
    first_arrive_record_should_come[people] = {}
    for date in first_arrive_record[people]:
        if date in date_people_should_come:
            first_arrive_record_should_come[people][date] = first_arrive_record[people][date]
```

```python
## the average coming time
arrival_time = []
for people in first_arrive_record_should_come:
    for date in first_arrive_record_should_come[people]:
        if first_arrive_record_should_come[people][date][1] == "A66 - 04" and first_arrive_record_should_come[people][date][0] < 13*3600:
            arrival_time.append(first_arrive_record_should_come[people][date][0])
            
print(arrival_time)
```

```python
def time_print(time):
    hour = str(int(time//3600))
    mint = str(int(time%3600//60))
    sec = str(int(time%3600%60))
    
    if len(hour) == 1:
        hour = '0' + hour
    if len(mint) == 1:
        mint = '0' + mint
    if len(sec) == 1:
        sec = '0' + sec
        
    print(f"The time is {hour}:{mint}:{sec}")
    
    
```

```python
average_time = sum(arrival_time)/len(arrival_time)
hour = average_time//3600
mint = average_time%3600//60
sec = average_time%3600%60
print(f"The average arrive time is 0{int(hour)}:0{int(mint)}:{int(sec)}")
```

```python
latest_arrival_time = max(arrival_time)
earliest_arrival_time = min(arrival_time)

time_print(latest_arrival_time)
time_print(earliest_arrival_time)
```

```python
arrive_time_people = {}
for people in first_arrive_record_should_come:
    arrive_time_people[people] = []
    for date in first_arrive_record_should_come[people]:
        arrive_time_people[people].append(first_arrive_record_should_come[people][date][0])
        
print(arrive_time_people)
```

```python
times_late = {}
for people in arrive_time_people:
    times_late[people] = 0
    for time in arrive_time_people[people]:
        if time > 9*3600:
            times_late[people] += 1
            
print(times_late)
```

```python
max_late_times = 0
for people in times_late:
    if times_late[people] > max_late_times:
        max_late_times = times_late[people]
  
late_most_people = []

for people in times_late:
    if times_late[people] == max_late_times:
        late_most_people.append(people)
        
print(late_most_people)
print(max_late_times)
```

```python
people_average_arrive_time = {}
for people in arrive_time_people:
    if len(arrive_time_people[people]) > 5:
        people_average_arrive_time[people] = sum(arrive_time_people[people])//len(arrive_time_people[people])
    
print(people_average_arrive_time)
```

```python
earliest_people = ""
earliest_time = 9*3600
latest_people = ""
latest_time = 0
for people in people_average_arrive_time:
    if people_average_arrive_time[people] > latest_time:
        latest_time = people_average_arrive_time[people]
        latest_people = people
    if people_average_arrive_time[people] < earliest_time:
        earliest_time = people_average_arrive_time[people]
        earliest_people = people
        
print(f"The earliest people is {earliest_people}")
print(f"The latest people is {latest_people}")
```

```python
#number of people arrived based on time preiod
number_people_arrive_period = {}
for i in range(24):
    number_people_arrive_period[i] = 0
    
for record in records_dic_list:
    number_people_arrive_period[record["time"]//3600] += 1
    
print(number_people_arrive_period)
time_list = []
period_list = []
for key in number_people_arrive_period:
    if number_people_arrive_period[key] != 0:
        time_list.append(number_people_arrive_period[key])
        period_list.append(key)
    

plt.figure()
plt.plot(period_list,time_list)
plt.figure()
plt.bar(range(len(time_list)), time_list, width=0.8, color='rgb', tick_label=period_list)
plt.show()
```
